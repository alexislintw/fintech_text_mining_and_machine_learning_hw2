# 驗證新聞對於股價波動的影響

## 讀取新聞發布n天之後的股價

```python
import os
import jieba
import numpy as np
import pandas as pd
from collections import Counter
from datetime import datetime ,timedelta
```

## 將新聞日期存取成datelist

```python
datelist = []
dir_path = './ctee_TSMC'
for file in os.listdir(dir_path):
    file_path = os.path.join(dir_path,file)
    datelist.append(file_path[-14:-4])
```

```python
st = pd.DataFrame(datelist)
st.to_csv('stockdate.csv')
```

## 讀取datelist區間股價

```python
stock_price = pd.read_csv('./TSMC_price_daily/TSMC2016010120190420.csv',encoding= "big5")
#stock_price = pd.read_csv('C:/Users/USER/github/HW1--data-mining/HW2/TSMC_price_daily/stock_price_2016120020190413.csv',encoding= "big5")
```

```python
stock_price.iloc[0:10,:]
```

## 將datelist 轉成datetime.dtype

```python
from datetime import datetime ,timedelta
datetimelist = []
for date in datelist :
    datetimelist.append(datetime.strptime(date,"%Y_%m_%d"))
datetimelist[:3]

datetimelist = sorted(datetimelist,reverse=True)
```

## 將stock_pirce_data 日期也轉換成datetime

```python
stock_list = stock_price['年月日'].tolist()
```

```python
stock_date = []
for date in stock_list :
    stock_date.append(datetime.strptime(date,"%Y/%m/%d"))
stock_price['年月日'] = stock_date
```

```python
dictstock = {}
index = 0
datetimelist01 = datetimelist
for i in stock_price['年月日'] :
    print('i=',i)
    for j in datetimelist01 :
        if (j >= i - timedelta(5) )   :
                print('j=' , j,stock_price['收盤價(元)'][index]) 
                dictstock[j]=stock_price['收盤價(元)'][index]
                datetimelist01 = datetimelist01[1:]
                break
    index += 1
```

### Check the len

```python
len(dictstock)
```

## 將datelist 延後5天(business)

```python
#from business_calendar import Calendar, MO, TU, WE, TH, FR
#cal = Calendar()
```

```python
datefive = []
for i in range(len(datetimelist)) :
    datefive.append(datetimelist[i] + timedelta(5))
```

```python
datefive[:5]
```

```python
len(datefive)
```

```python
dictstock_five = {}
index = 0
datefive01 = datefive
for i in stock_price['年月日'] :
    print('i=',i)
    for news in datefive01 :
        if (news >= i - timedelta(5) )   :
            print('news=' , news,stock_price['收盤價(元)'][index]) 
            dictstock_five[news]=(stock_price['收盤價(元)'][index] + stock_price['收盤價(元)'][index+1] + stock_price['收盤價(元)'][index+2]) / 3
            datefive01 = datefive01[1:]
            break
    index += 1
```

```python
len(dictstock_five)
```

# check each news price and concate to a Dateframe

```python
df_stock= pd.DataFrame.from_dict(dictstock,orient='index',columns = ['Now_price'])
df_stock_after = pd.DataFrame.from_dict(dictstock_five,orient='index',columns = ['After_price'])
```

```python
df_stock.reset_index(inplace=True)
df_stock_after.reset_index(inplace=True)
df_stock.columns = ['Date','Now_price']
df_stock_after.columns = ['Date_after','After_price']
```

```python
news_stock_prices = pd.concat([df_stock,df_stock_after],axis=1)
```
